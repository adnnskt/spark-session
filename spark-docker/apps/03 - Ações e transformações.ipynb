```python

from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Ações e transformações").getOrCreate() # inicializa o spark
```

```python
from pyspark.sql.types import *

arqschema = " id INT, nome STRING, status STRING, cidade STRING, vendas INT, data STRING"

despachantes = spark.read.csv("despachantes.csv", header = False, schema = arqschema)
despachantes.show()
```

```python
#Exemplo orderBy

from pyspark.sql import functions as Func

despachantes.orderBy(Func.col("cidade").desc(), Func.col("vendas").desc()).show() 
```

```python
#Exemplo de agregação com soma de vendas e group By

despachantes.groupBy("cidade").agg(sum("vendas")).show() # forma simples

despachantes.groupBy("cidade").agg(sum("vendas")).orderBy(Func.col("sum(vendas)")).show() # adicionando ordenação
```

```python
# Utilizando Filter para filtrar dados

despachantes.filter(Func.col("id") == 2).show()
```
