```python
from pyspark.sql import SparkSession

spark  = SparkSession.builder.appName("Exportando dados").getOrCreate()
```

```python
despachantes = spark.read.csv("despachantes.csv")
despachantes.show()
```

```python
# persistindo dados do DF no diretório intalke 

despachantes.write.format("parquet").save("intake/despachantes")
```

```python
despachantes.write.format("json").save("intake/despachantesJson")
despachantes.write.format("csv").save("intake/despachantesCsv")
```
