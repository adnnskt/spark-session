```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *

spark = SparkSession.builder.appName("spark sql").getOrCreate()

```

```python
spark.sql("show databases").show()
```

```python
spark.sql("CREATE DATABASE DESP") 
```

```python
spark.sql("use desp")
```

```python
#lendo dados despachantes

arqschema = "id INT,nome STRING, status STRING, cidade STRING, vendas INT, data STRING"

despachantes = spark.read.csv("dados/despachantes.csv", header = False, schema = arqschema)
despachantes.write.saveAsTable("despachantes")
```

```python
spark.sql("show tables").show()
spark.sql("select * from despachantes").show()
```

```python

#salvando dados em .parquet para uso de tabelas não gerenciadas
despachantes.write.format("parquet").save("dados/despparquet")
```
