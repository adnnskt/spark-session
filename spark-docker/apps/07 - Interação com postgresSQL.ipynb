```python
from pyspark.sql import SparkSession 
from pyspark.sql.types import *
spark = SparkSession.builder.appName("Integração postgresSQL").getOrCreate()
```

```python
# Instale o pacote se necessário (descomente se precisar)
#!pip install psycopg2-binary
import psycopg2
conn = psycopg2.connect(
    host='postgres-db',  # ou o nome do serviço no docker-compose / container_name
    port=5432,        # porta padrão do PostgreSQL
    database='sparkdb',
    user='sparkuser',
    password='sparkpass'
 )
print('Conexão estabelecida com PostgreSQL')
```

```python
resumo = spark.read.format("jdbc").option("url", "jdbc:postgresql://localhost:5432/vendas").option("dbtable", "vendas").option("user", "sparkuser").option("password", "sparkpass").option("driver", "org.postgresql.Driver").load()
resumo.show()
```
