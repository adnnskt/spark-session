```python
from pyspark.sql import SparkSession 
from pyspark.sql.types import *

spark = SparkSession.builder.appName("Interação mongodb").getOrCreate()
```

```python
# Configurações de conexão com o MongoDB
spark.conf.set("spark.mongodb.input.uri", "mongodb://mongo:27017/posts.post")
spark.conf.set("spark.mongodb.output.uri", "mongodb://mongo:27017/posts.post")
```

```python
#consulta para ler os dados do MongoDB
posts = spark.read.format("mongo") \
    .option("uri", "mongodb://mongo:27017/posts.post") \
    .load()

posts.show()
```

```python
#Salvando dados no MongoDB
posts.write.format("mongo") \
    .option("uri", "mongodb://mongo:27017/posts2.post") \
    .save()
```
